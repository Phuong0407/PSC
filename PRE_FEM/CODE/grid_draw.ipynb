```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
H2 = 10       # Height scaling parameter
L1 = 4        # Starting length parameter
L2 = 3        # Ending length parameter
N = 20        # Number of divisions in the i-direction
M = 20      # Number of divisions in the j-direction

# Initialize arrays for X and Y
X = np.zeros((N+1, M+1))
Y = np.zeros((N+1, M+1))

# Generate the grid based on the given equations
for i in range(N+1):  # i ranges from 0 to N
    for j in range(M+1):  # j ranges from 0 to M
        X[i, j] = (H2 / N) * i
        Y[i, j] = ((L2 - L1) / M) * (j * i / N) + (L1 / M) * j

# Plot the grid
plt.figure(figsize=(8, 8))

# Draw horizontal and vertical grid lines
for i in range(N+1):
    plt.plot(X[i, :], Y[i, :], 'k-', lw=1)  # Horizontal lines
for j in range(M+1):
    plt.plot(X[:, j], Y[:, j], 'k-', lw=1)  # Vertical lines

# Scatter points
plt.scatter(X, Y, color='red', s=10, label='Grid Points')

# Add labels and title
plt.title("Parametric Grid Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.axis('equal')  # Equal scaling for X and Y axes
plt.show()
```
